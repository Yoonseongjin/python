```python
import requests
import matplotlib.pyplot as plt

url = 'http://apis.data.go.kr/B552061/lgStat/getRestLgStat'
params = {
    'serviceKey': 'ZaVwFFRsweQREQo8+971pc4gigS7yKOotJVXqc5cnBRQAshsPIquWDuCBeHV00+7YjkhC0hv1RGHZ02S3QXi9A==',
    'searchYearCd': '2019',
    'siDo': '1100',
    'guGun': '1116',
    'type': 'json',
    'numOfRows': '10',
    'pageNo': '1'
}

response = requests.get(url, params=params)
data = response.json()

# 데이터 추출
item_list = data['items']['item']

# 데이터 처리
x_labels = []
tot_acc_cnt_list = []
dth_dnv_cnt_list = []
injpsn_cnt_list = []

for item in item_list:
    x_labels.append(item['acc_cl_nm'])
    tot_acc_cnt_list.append(int(item['tot_acc_cnt']))
    dth_dnv_cnt_list.append(int(item['tot_dth_dnv_cnt']))
    injpsn_cnt_list.append(int(item['tot_injpsn_cnt']))

# 그래프 그리기
width = 0.2
x = range(len(x_labels))

plt.bar(x, tot_acc_cnt_list, width=width, label='전체 사고')
plt.bar([i + width for i in x], dth_dnv_cnt_list, width=width, label='사망자 수')
plt.bar([i + 2 * width for i in x], injpsn_cnt_list, width=width, label='부상자 수')

plt.xlabel('')
plt.ylabel('수')
plt.title('2019년도 교통사고 통계')
plt.xticks([i + width for i in x], x_labels, rotation=45)
plt.legend()
plt.show()

```
